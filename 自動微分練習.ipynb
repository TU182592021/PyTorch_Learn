```python
import torch
```

$y=3x^{3}$

```python
x=torch.tensor([2],dtype=torch.float32,requires_grad=True)
```

```python
y=3*x**3
```

```python
y.backward()
```

```python
x.grad
```

```python
x1=torch.tensor([4],dtype=torch.float32,requires_grad=True)
x2=torch.tensor([2],dtype=torch.float32,requires_grad=True)
```

```python
y=2*x1**2+3*x2**3
```

```python
y.backward()
```

```python
x1.grad
```

```python
x2.grad
```
